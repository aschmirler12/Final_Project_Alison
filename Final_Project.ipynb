### b. String columns (Book-Title, Book-Author, Publisher) 

Function that creates a new dataframe containins the rows that only have ascii encoding.

```python
import re

def clear_column(col:str,df:pd.DataFrame) -> pd.DataFrame:
  """ returns a dataframe containing only rows that are ascii encoded"""
  ascii_rows = []

  for index, title in df[col].items():
      result = chardet.detect(str(title).encode())
      if result['encoding'] == 'ascii':
          ascii_rows.append(index)

  new_df = df.loc[ascii_rows]
  return new_df
```

```python
df = clear_column('Title',df)
df = clear_column('Author',df)
df = clear_column('Publisher',df)
df
```

```python
pattern = r'&(\w)*(;)*|[\\"]'
df['Publisher'] = df['Publisher'].str.replace(pattern, '&')
df['Title'] = df['Title'].str.replace(pattern, '')
df['Author'] = df['Author'].str.replace('[^a-zA-Z\s\.]', '')
df['Author'] = df['Author'].str.title()
df
```

# Step 2: Add description column

```python
import requests
import json
import time

url = "https://www.googleapis.com/books/v1/volumes"
api_key = input("API Key:")

for i in df_top['ISBN']:

  params = {
      'q': 'isbn:' + i,
      'key': api_key
  }

  response = requests.get(url, params=params)
  data = response.json()
  try:
    if data['totalItems'] == 0:
      df_top = df_top.drop(df_top.loc[df_top['ISBN'] == i].index)
    else:
      description = data['items'][0]['volumeInfo'].get('description', '')
      df_top.loc[df_top['ISBN'] == i, 'Description'] = description
  except KeyError:
    df_top = df_top.drop(df_top.loc[df_top['ISBN'] == i].index)
  time.sleep(0.05)

df_top
```

```python
df_top.reset_index(drop=True, inplace=True)
df_top.rename(columns={'index': 'ID'}, inplace=True)
df_top
```

# Step 3: Make an Author table

### 3.1. Create table and add description and image columns




```python
df_authors = df_top['Author'].reset_index()
df_authors['Description'] = ''
df_authors['Image'] = ''
df_authors
```

### 3.2. Use Wikipedia API to extract description and Image for each author

```python
import requests

for i in df_authors['Author']:
    base_url = 'https://en.wikipedia.org/w/api.php'
    params = {
        'action': 'query',
        'format': 'json',
        'prop': 'extracts|pageimages',
        'exintro': '',
        'explaintext': '',
        'titles': i
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    # Get the page ID of the author
    page_id = next(iter(data['query']['pages'].keys()))

    try:
      desc = data['query']['pages'][page_id]['extract']
      img = data['query']['pages'][page_id]['thumbnail']['source']
      df_authors.loc[df_authors['Author'] == i, 'Description'] = desc
      df_authors.loc[df_authors['Author'] == i, 'Image'] = img
    except:
      continue
```

```python
df_authors.replace('',np.nan,inplace=True)
df_authors.info()
```

```python
df_authors
```

### 3.3. Add a column showing how many books the author wrote

```python
df_auth_count=df_top.groupby('Author')['Title'].count().reset_index()
df_authors = pd.merge(df_auth_count,df_authors,on='Author')[['index','Author','Description','Image','Title']]
df_authors.rename(columns={'index':'ID','Title':'No_Books'},inplace=True)
```

### 4.1 Function to implement the search bar

```python
def search_bar(search,df):
  df = df.astype(str)

  filtered_df = df[df.applymap(lambda x: search in str(x)).any(axis=1)]
  return filtered_df
```

```python
while True:
  option = input("""Search for:
  1. Authors
  2. Books
  3. Exit
  Option := """)

  dict_options = {
      "1": df_authors,
      "2": df_top
  }

  if option in dict_options:
    search = input("Search: ")
    display(search_bar(search,dict_options[option]))
  elif option == "3":
    break
  else:
    print("No valid option. ")
```
